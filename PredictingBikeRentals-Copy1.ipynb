```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
bike_rentals = pd.read_csv('bike_rental_hour.csv')
bike_rentals.head(5)
```

```python
np.histogram(bike_rentals['cnt'])
```

```python
%matplotlib inline

plt.hist(bike_rentals['cnt'], bins=11)
plt.show
```

```python
bike_rentals.corr()['cnt']
```

```python
def assign_label(nmbr):
    if nmbr in range(6,13):
        return 1
    elif nmbr in range(12, 19):
        return 2
    elif nmbr in range(18, 25):
        return 3
    elif nmbr in range(0, 7):
        return 4

bike_rentals['time_label'] = bike_rentals['hr'].apply(assign_label)
bike_rentals.head(25)
```

Based on our observation of the count column, it seems most rentals consist of less than 200 people.So we'll use that as our error metric.

```python
train = bike_rentals.sample(frac=0.80, replace=False, axis=0)
test = bike_rentals.loc[~bike_rentals.index.isin(train.index)]
print(train.shape, test.shape)
```

```python
predictors = list(train.columns)
predictors.remove('cnt')
predictors.remove('casual')
predictors.remove('dteday')
predictors.remove('registered')

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

lr = LinearRegression()
lr.fit(train[predictors], train['cnt'])

test_predictions = lr.predict(test[predictors])
mse = mean_squared_error(test['cnt'], test_predictions)
rmse = np.sqrt(mse)
rmse
```

```python
test['cnt'].mean()
```

```python
test['cnt'].std()
```

The rmse is within the std, however there is a lot of variance since this is a high std.

```python
from sklearn.tree import DecisionTreeClassifier

dc = DecisionTreeClassifier(min_samples_leaf=1)
dc.fit(train[predictors], train['cnt'])

test_predictions = dc.predict(test[predictors])
mse = mean_squared_error(test['cnt'], test_predictions)
rmse = np.sqrt(mse)
rmse
```

The decision tree model is much better than that of linear regression. Also, the rmse increased with increasing min_samples_leaf. Hence the default of min_samples_lef=1 gived the best model .

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(min_samples_leaf=4)
rf.fit(train[predictors], train['cnt'])

test_predictions = rf.predict(test[predictors])
mse = mean_squared_error(test['cnt'], test_predictions)
rmse = np.sqrt(mse)
rmse
```

The random forest regressor gave the best model. Due to randomization, the rmse varies slightly between each run for min_samples_leaf values, for the same value and for values that are close together.
